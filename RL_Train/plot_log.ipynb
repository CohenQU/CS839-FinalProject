```python
import numpy as np
from matplotlib import pyplot as plt

avgs = []
path = "./logs/td3/Ant-v3_23/evaluations.npz"
with np.load(path) as data:
    t = data['timesteps']
    r = data['results']
    avg = np.average(r, axis=1)
    avgs.append(avg)

t= np.array(t.resize(1, 40))
plt.plot(t, avgs)
plt.show()
```
